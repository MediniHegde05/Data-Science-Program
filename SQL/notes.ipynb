## Data Warehouse <br>
1. It is used by a company to store all its data in databases. <br>
2. Data is stored such that its retrieval and management is easy.

a data warehouse is characterised by the following:<br>

Subject oriented: Built for a specific purpose<br>
Integrated: Collated from multiple sources into one form<br>
Non-volatile: Data does not change over time<br>
Time variant: Capable of capturing information over time<br>

Data is captured from multiple sources and stored in Excel files or in databases, such as Oracle.<br>
1. The Extract, Transform, Load (ETL) process is performed on the captured data in the following manner:<br>
2. Extract: Extracting the information from the data sources<br>
3. Transform: Transforming the values to the required format, such as converting rupees to dollars<br>
4. Load: Loading the data into a central repository, which is none other than a data warehouse<br>
5. The stored data is either connected to Online Analytical Processing (OLAP) for further processing or used for creating reports in tools, such as Tableau or PowerBI.<br>

### A DIMENSION MODELLING <br>
A dimension model is a database structure technique and is used to optimize the database for the fast retrieval of data. The dimensional model contains two entities: Facts and Dimensions.  <br>
<B>Facts</B> are the numerical data and <B>imensions </B> are the metadata (that is, data explaining some other data) attached to the fact variables. Both facts and dimensions are equally important for generating actionable insights from a data set.
<B> ERD </B> an Entity Relationship Diagram (ERD) is a representation of the tables in a database. It gives us an idea of how different entities are connected. ERDs are extremely useful to get an overall idea of a database in very less time. <br>
 <br>
Here is a short summary of what each notation in an ERD signifies:<br>

Fields marked with a star are the primary keys for each table.<br>
A line connecting two tables indicates that a relationship exists between them.<br>
A line connecting a table to itself indicates that the table references itself.<br>
There are four types of relationships that exist in an ERD:<br>

One to One: One to one type of relationship exists, when a single instance of an entity is related to only a single instance of another entity.<br>
One to Many: One to many type of relationship exists, when a single instance of an entity is related to more than one instance of another entity.<br>
Many to One: Many to one type of relationship exists, when more than one instance of an entity is related to only one instance of another entity.<br>
Many to Many: Many to many type of relationship exists, when more than one instance of an entity is related to more than one instance of another entity.<br>

### Star and Snowflake Schemas <br>
A schema is an outline of the entire data model which shows how different data sets are connected and how the different attributes of each data set are used for the database design.<br>

#### STAR SCHEMA <br>
A star schema is a type of schema design that has one fact table connected to multiple dimension tables.<br>
Easy to understand<br>
Easy to implement<br>
Efficient in terms of querying because a maximum of only one join is required<br>

#### SNOWFLAKE SCHEMA <br>
A snowflake schema is called so because the dimension tables surrounding the fact table can branch off into more dimension tables.<br>
Snowflake schemas are particularly useful when it becomes difficult to manage the size of the dimensional tables

#### OLAP vs OLTP <br>
OLTP(Online transcation processing) is used for day-to-day transactions, whereas OLAP(online analytical processing) is used for analytical purposes. 
OLTP is used by end users for business transcations and OLAP is used by CEO,Managers etc for decision making <br>
no technology used in OLTP meaning no ETL process, only DBMS and API, Whereas in OLAP ETL process is used
OLATP Database design is based on integrity issues and normalisation no star schema and nature of retreive is CRUD operation, where as in OLAP is based on dimensional modelling, creation of star schema nature of access is retreive 

#### ENTITY CONSTRAINTS <br>
constraints are the rules used to restrict the values that can be stored in the column of database <br>
Types of entity constraints:<br>
1. unique :example employee ID should be unique in employee table <br>
2. null constraint: used to determine if a column can have null value or not <br>
3. Primary Key: Determines the column that uniquely identifies a table.<br>

#### Referential Constraints <br>
These are used to restrict the values that are taken by a column in one table based on the values that exist in another table.<br>
A referential constraint is a rule between two tables. According to this rule, the value that appears as a foreign key in a table is valid only if it also appears as a primary key in the table to which it refers.<br>
A given table has only one primary key but it can have multiple foreign keys. <br>
A primary key identifies each row in a table uniquely, whereas a foreign key uniquely identifies a row in the same table or in another table.

#### Semantic Constraints <br>
These impose restrictions on the values of column <br>
Using a semantic constraint ensures that we do not get incorrect data for any row in the database.<br>
For example, a row with an Indian phone number not having 10 digits after the country code would not be allowed to enter the database.
