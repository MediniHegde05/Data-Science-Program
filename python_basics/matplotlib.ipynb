### Facts and Dimensions <BR>
Facts and dimensions are different types of variables that help you interpret data better. Facts are numerical data, and dimensions are metadata. Metadata explains the additional information associated with the factual variable. <br>
 For example, in a data set about the height of students in a class, the height of the students would be a fact variable, whereas the gender of the students would be a dimensional variable

### Bar Graph

```python
import numpy as np 
import matplotlib.pyplot as plt 
product_catergory = np.array(['Furniture', 'Technollogy', 'Office Supplies'])
sales = np.array ([4110451.90, 4744557.50, 3787492.52])
plt.bar(product_catergory,sales, width = 0.5, align = 'center', edgecolor = "Orange", color = "green")

# Adding and formating title
plt.title("Sales Across Product Categories", fontdict= {'fontsize': 20,'fontweight':5, 'color': 'Black'})

# Labelling axis
plt.xlabel("Product Categories", fontdict= {'fontsize':12,'fontweight':5, 'color': 'Brown'})
plt.ylabel("Sales", fontdict= {'fontsize':12,'fontweight':5, 'color': 'Brown'})

# Modifying ticks to show in million dollars
ticks =  np.arange(0,6000000, 1000000)
labels = ["{}M".format(i //1000000) for i in ticks]
plt.yticks(ticks, labels)
plt.show()


```

A line graph is used to present continuous time-dependent data. It accurately depicts the trend of a variable over a specified time period. Lets watch the next video to learn how to plot a line chart using the Matplotlib library.
