```python

Spotify_data['Spotify Streams'] = Spotify_data['Spotify Streams'].fillna(0).astype(int)
```

```python
top_song = Spotify_data.sort_values(by='Spotify Streams', ascending=False).head(10)
top_song
```

```python
spotify_drop = Spotify_data.drop([ 'ISRC','YouTube Views', 'YouTube Likes', 'TikTok Posts',
       'TikTok Likes', 'TikTok Views', 'YouTube Playlist Reach',
       'Apple Music Playlist Count', 'AirPlay Spins', 'SiriusXM Spins',
       'Deezer Playlist Count', 'Deezer Playlist Reach',
       'Amazon Playlist Count', 'Pandora Streams', 'Pandora Track Stations',
       'Soundcloud Streams', 'Shazam Counts', 'TIDAL Popularity',
       'Explicit Track'
             ],axis='columns')
spotify_drop
```

```python
Spotify_data['Track'].isna()
```

```python
Spotify_data.dropna(subset=['Track'], inplace=True)
Spotify_data.dropna(subset=['Artist'], inplace=True)
Spotify_data.dropna(subset=['Spotify Streams'], inplace=True)
print(Spotify_data.isna().sum())
```

```python
songcount_by_artist = (Spotify_data.groupby('Artist')['Track'].count().reset_index(name='Track').sort_values(by='Track',ascending = False).head(10))
songcount_by_artist= songcount_by_artist.reset_index(drop=True)
songcount_by_artist
```

```python
import numpy as np 
import matplotlib.pyplot as plt 
most_tracks_artist = songcount_by_artist.iloc[0]['Artist']
most_tracks_count = songcount_by_artist.iloc[0]['Track']
colors = plt.cm.get_cmap('Set2', len(most_tracks_artist))
bar_colors = colors(np.arange(len(most_tracks_artist)))
plt.bar(songcount_by_artist['Artist'], songcount_by_artist['Track'],align = 'center', color= bar_colors)
plt.xlabel('Artist')
plt.ylabel('Track Count')
plt.title('Artists with Most Tracks (Top 10)')
plt.xticks(rotation=45, ha='right') 
plt.tight_layout()
plt.show()
```

songs released in past six months that have the highest spotify streams

```python

Spotify_data['Release Date'] = pd.to_datetime(Spotify_data['Release Date'], format='%d-%m-%Y', errors='coerce')
six_months = pd.Timestamp.now() - pd.DateOffset(months=6)
six_months
```

```python
Spotify_data.drop_duplicates(subset=['Track', 'Album Name', 'Artist'], keep= "last")
```

```python
past_six_months = Spotify_data[Spotify_data['Release Date'] >= six_months]
recent_songs_sorted = past_six_months.sort_values(by='Spotify Streams', ascending=False)
Spotify_data.drop_duplicates(subset=['Track', 'Album Name'], keep= "last")
top_songs = recent_songs_sorted.head(20)
top_songs
```

```python
import numpy as np 
import matplotlib.pyplot as plt 
colors = np.random.rand(len(top_songs), 3)
plt.bar(top_songs['Track'], top_songs['Spotify Streams'], color=colors)
plt.xlabel('Track')
plt.ylabel('Spotify Streams')
plt.title('Spotify Most streamed Song in Past 6 months(Top 20)')
ticks = np.arange(0, 2045512421, 100000000)  # Adjust the range and step size as needed
labels = ["{}M".format(int(i // 1000000)) for i in ticks]
plt.yticks(ticks, labels)
plt.yticks(ticks, labels)
plt.xticks(rotation=45, ha='right') 

plt.show()
```
