```python
import pandas as pd
Bank_data = pd.read_csv("C:\\Users\\medin\\Downloads\\bank_marketing_updated_v1.csv", header=0, encoding='latin1')
Bank_data
Bank_data.head()
```

Fixing the Rows and Columns

```python
Bank_data.columns = Bank_data.iloc[1]
Bank_data.drop(Bank_data.index[1])
Bank_data = Bank_data.drop(0)
Bank_data = Bank_data.drop(1)
Bank_data
```

```python
#print the information of variables to check their data types.
Bank_data.info()
#numeric data: salary, balance, duration and age
```

```python
Bank_data["salary"] = Bank_data["salary"].fillna(0).astype(int)
Bank_data["balance"] = Bank_data["balance"].fillna(0).astype(int)
Bank_data["age"] = Bank_data["age"].fillna(0).astype(int)

Bank_data.info()
```
